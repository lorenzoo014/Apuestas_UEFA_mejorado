Se va a realizar un data cleaning de los csv de equipos.csv y partidos.csv los cuales son imprescindibles a la hora de predecir las cuotas el cual es el
objetivo final de la practica


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('docs/equipos.csv')
```

```python
print(data.head())
```

```python
columna1 = data['Puntuaje']
media = np.mean(columna1)
desviacion_estandar = np.std(columna1)
print("Media: ", media)
print("Desviación estándar: ", desviacion_estandar)
```

```python
columna3 = data['Pais']
plt.hist(columna3, bins=10)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de columna3')
plt.show()
```

```python
columna4 = data['Equipo']
columna5 = data['Puntuaje']
plt.scatter(columna4, columna5)
plt.xlabel('Equipo')
plt.ylabel('Puntuaje')
plt.title('Gráfico de dispersión entre Equipo y Puntuaje')
plt.show()
```

```python
data = pd.read_csv('docs/Partiditos.csv')
```

```python
print(data.head())
```

```python
columna4 = data['equipo_local']
columna5 = data['equipo_visistante']
plt.scatter(columna4, columna5)
plt.xlabel('equipo_local')
plt.ylabel('equipo_visistante')
plt.title('Gráfico de dispersión entre equipo_local y equipo_visistante')
plt.show()
```

```python
# Grafica 2: Gráfico de barras de una columna
plt.bar(data['equipo_local'], data['equipo_visistante'])
plt.title('Gráfico de barras')
plt.xlabel('equipo_local')
plt.ylabel('equipo_visistante')
plt.show()
```
